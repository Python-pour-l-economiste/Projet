2) Nettoyage du dataframe df_détail_étape

Le nettoyage est résolument similaire que celui du dataframe df_synthèse. On commence d'abord par enlever les données manquantes et les colonnes redondantes du dataframe

```python
df_détail_etape = df_détail_etape.dropna()
df_détail_etape.drop(["Code CIQUAL", "LCI Name", "Groupe d'aliment", "Nom et code","DQR - P", "DQR - TiR", "DQR - GR", "DQR - TeR"], axis='columns', inplace=True)
df_détail_etape
```

On remarque que les indicateurs précédemment triés se séparent entre six catégories correspondant aux six étapes du cycle de vie des aliments étudiés.
Afin de pouvoir exploité le dataframe, il serait pratique de regrouper les indicateurs par étape du cycle de vie afin de cibler l'impact de chaque étape sur l'environnement.
Nous crééons ainsi des indicateurs synthétiques qui réalisent la moyenne pondérée de l'ensemble des indicateurs concernant cette même étape. 
Nous ne conservons ainsi que le dataframe contenant les éléments essentiels de présentation des aliments ainsi que ces indicateurs synthétiques.

```python
df_détail_etape["Score unique EF (mPt / kg de produit)"] = (df_détail_etape["Score unique EF (mPt / kg de produit) - Agriculture"] + df_détail_etape["Score unique EF (mPt / kg de produit) - Transformation"] + df_détail_etape["Score unique EF (mPt / kg de produit) - Emballage"] + df_détail_etape["Score unique EF (mPt / kg de produit) - Transport"] + df_détail_etape["Score unique EF (mPt / kg de produit) - Supermarché et distribution"] + df_détail_etape["Score unique EF (mPt / kg de produit) - Consommation"])/6
df_détail_etape_synthese = df_détail_etape[["Sous-groupe d'aliment",'Nom du Produit en Français',"DQR","Score unique EF (mPt / kg de produit)", "Score unique EF (mPt / kg de produit) - Agriculture", "Score unique EF (mPt / kg de produit) - Transformation", "Score unique EF (mPt / kg de produit) - Emballage", "Score unique EF (mPt / kg de produit) - Transport","Score unique EF (mPt / kg de produit) - Supermarché et distribution", "Score unique EF (mPt / kg de produit) - Consommation"]]
df_détail_etape_synthese
```

A nouveau, on ne garde que les données qui ont un DQR (data quality ratio) plus faible que 3 (voir supra pour la justification du seuil).

```python
df_détail_etape_synthese = df_détail_etape_synthese[df_détail_etape_synthese["DQR"]<=3].copy()
df_détail_etape_synthese
```

De même, nous supprimmons les doublons dans les noms des produits afin de conserver qu'une seule ligne par produit de même appelation (on s'assure que chaque produit du tableau a une valeur différente de score unique).

```python
df_détail_etape_synthese = df_détail_etape_synthese.drop_duplicates(subset = ['Score unique EF (mPt / kg de produit)'])
df_détail_etape_synthese
```

Finalement, nous classons comme précédemment les données en fonction du sous-groupe d'aliment.

```python
df_détail_etape_synthese = df_détail_etape_synthese.sort_values("Sous-groupe d'aliment", ascending = True)
df_détail_etape_synthese
```

```python
df_détail_etape_synthese = df_détail_etape_synthese.reset_index(drop=True)
df_détail_etape_synthese
```

```python
X=pd.DataFrame(np.c_[df_synthese['Changement climatique (kg CO2 eq/kg de produit)'],df_synthese["Appauvrissement de la couche d'ozone (E-06 kg CVC11 eq/kg de produit)"], df_synthese["Rayonnements ionisants (kBq U-235 eq/kg de produit)"], df_synthese["Acidification terrestre et eaux douces (mol H+ eq/kg de produit)"], df_synthese["Utilisation du sol (Pt/kg de produit)"]])
Y=df_synthese['Score unique EF (mPt/kg de produit)']

from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.3, random_state=5)
 
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

lmodellineaire = LinearRegression()
lmodellineaire.fit(X_train, Y_train)

from sklearn.metrics import r2_score

y_train_predict = lmodellineaire.predict(X_train)
rmse = (np.sqrt(mean_squared_error(Y_train, y_train_predict)))
r2 = r2_score(Y_train, y_train_predict)
print('La performance du modèle sur la base dapprentissage')
print('--------------------------------------')
print('Lerreur quadratique moyenne est {}'.format(rmse))
print('le score R2 est {}'.format(r2))
print('\n')
 
y_test_predict = lmodellineaire.predict(X_test)
rmse = (np.sqrt(mean_squared_error(Y_test, y_test_predict)))
r2 = r2_score(Y_test, y_test_predict)
print('La performance du modèle sur la base de test')
print('--------------------------------------')
print('Lerreur quadratique moyenne est {}'.format(rmse))
print('le score R2 est {}'.format(r2))
 
a = lmodellineaire.coef_
print(a)
```
